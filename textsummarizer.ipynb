```python
import pandas as pd
nba = pd.read_csv("train.csv")
nba
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize

nltk.download('punkt')
nltk.download('stopwords')

def preprocess_text(text):
    # Remove special characters and extra whitespaces
    text = re.sub(r'\s+', ' ', text)  # Replace multiple whitespaces with a single whitespace
    text = re.sub(r'[^\w\s]', '', text)  # Remove non-alphanumeric characters

    # Tokenize text into words
    words = word_tokenize(text.lower())  # Convert text to lowercase during tokenization

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in words if word not in stop_words]

    return filtered_words
```

```python
nba['preprocessed_text'] = nba['article'].apply(preprocess_text)
print(nba['preprocessed_text'].head())
```
